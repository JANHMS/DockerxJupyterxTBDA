```
def get_time_window (low, delta):

    B = datetime.strptime(low, '%d/%m/%y %H:%M:%S')
    E = B + timedelta(days=delta)

    return (round(datetime.timestamp(B)*1000),round(datetime.timestamp(E)*1000))
```

```
def sql_get_alerts(low,delta):
    B,E=get_time_window(low,delta)
    sql_command="SELECT  date, value FROM public.variable_log_string  WHERE date BETWEEN {0} and {1} GROUP BY value, date ".format(B,E)
    return sql_command
```

### We want to aggregate the errors, which happen in the same day first

```
aggregation_functions = {'errors': 'sum', 'hour': 'first', 'day': 'first'}
df_errors_aggregae = inputs.groupby(inputs['day']).aggregate(aggregation_functions)
df_errors_aggregae_sorted_per_hour_day = df_errors_aggregae.sort_values(['errors'], ascending=False)
df_errors_aggregae_sorted_per_hour_day
```

### We can see most errors are happending at 09-11, so in the morning time.

```
# currently hour is the index, but we need to change that to sort the df by hour ascending
df_errors_aggregae_sorted_per_hour_day['index_col'] = df_errors_aggregae_sorted_per_hour_day.index

df_errors_aggregae_sorted_per_day = df_errors_aggregae_sorted_per_hour_day.sort_values(['index_col'], ascending=True)
df_errors_aggregae_sorted_per_day
```

```
import plotly.express as px

fig = px.line(df_errors_aggregae_sorted_per_day, x="day", y="errors", title='Errors per day at specific day')
fig.show()
```

First approach using a lambda and groupby

```
df_errors_lamda_sorted_per_hour_per_day = inputs.groupby(['day', 'hour'])["errors"].apply(lambda x : x.astype(int).sum())
df_errors_lamda_sorted_per_hour_per_day
```

Second approach using a aggregate

```
df_errors_aggregae_sorted_per_hour_per_day = inputs.groupby(['day','hour']).agg({'errors':'sum'}).reset_index()
df_errors_aggregae_sorted_per_hour_per_day
```
